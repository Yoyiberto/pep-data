```python
import torch
from transformers import AutoTokenizer
from unsloth import FastLanguageModel

# This is the path to your uploaded model directory.
# Make sure it matches the name of the folder you uploaded.
model_dir = "/content/drive/MyDrive/2025/PEP/finetuning_2/mg_verilog_lora_model"

# Load the model from the specified directory.
# Unsloth will automatically handle loading the base model and applying your
# LoRA adapter on top.
model, tokenizer = FastLanguageModel.from_pretrained(
    model_name = model_dir,       # Your model adapter directory
    max_seq_length = 2048,        # Can be any value, Unsloth handles scaling
    dtype = None,                 # Autodetects for your GPU
    load_in_4bit = True,          # Load in 4-bit for memory efficiency
)

# Enable inference mode for faster generation.
FastLanguageModel.for_inference(model)

```

```python
# ==============================================================================
# CELL 4: Run inference
# ==============================================================================
# Now you can test your fine-tuned model with a prompt.
# Since your model is fine-tuned on Verilog data, let's use a relevant prompt.

prompt = [
    {"role": "user", "content": "Write in VHDL the following: For each bit in an 8-bit vector, detect when the input signal changesfrom 0 in one clock cycle to 1 the next (similar to positive edgedetection). The output bit should be set the cycle after a 0 to 1transition occurs."}
]

# Prepare the prompt for the model
inputs = tokenizer.apply_chat_template(
    prompt,
    tokenize = True,
    add_generation_prompt = True,  # Required for generation
    return_tensors = "pt",
).to("cuda")

# Generate a response from the model
outputs = model.generate(input_ids = inputs, max_new_tokens = 512, use_cache = True)
response = tokenizer.batch_decode(outputs)

# Print the generated response
print(response[0])

# You can try another prompt to see how it responds.
prompt_2 = [
    {"role": "user", "content": "Create a module that implements an XNOR gate."}
]

inputs_2 = tokenizer.apply_chat_template(
    prompt_2,
    tokenize = True,
    add_generation_prompt = True,
    return_tensors = "pt",
).to("cuda")

outputs_2 = model.generate(input_ids = inputs_2, max_new_tokens = 512, use_cache = True)
response_2 = tokenizer.batch_decode(outputs_2)

print("\n" + "="*50 + "\n")
print(response_2[0])
```
